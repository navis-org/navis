```python
import pymaid
import matplotlib.pyplot as plt

# Connect to CATMAID
rm = pymaid.CatmaidInstance( 'www.your.catmaid-server.org',
                             'HTTP_USER' ,
                             'HTTP_PASSWORD',
                             'TOKEN' )

# Get some neurons
nl = pymaid.get_neurons(['57311','27295'])
```

```python
fig, ax = pymaid.plot2d(nl, method='2d',
                        connectors=False,
                        linewidth=1.5,
                        depth_coloring=True)

plt.show()
```

```python
fig, ax = pymaid.plot2d(nl, method='3d',
                        connectors=False,
                        linewidth=1.5,
                        depth_coloring=True)

# Rotate to top view
ax.elev = 90

# Zoom in a bit
ax.dist = 6

plt.show()
```
